```python
#测试概率中的各种分布
#参考http://python.jobbole.com/81321/
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
#stats.binom计算二项分布
n = 5 #进行试验的次数
p = 0.5 #试验成功的概率
k = np.arange(0, n+1)
rv = stats.binom(n, p) #计算二项分布
print(rv.pmf(k)) #打印出事件成功 0, 1, 2, ...次的概率

plt.plot(k, rv.pmf(k), 'o-')
plt.title('Binomial: n=%i, p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Number of successes')
plt.ylabel('Probability of successes', fontsize=15)
plt.show()
```

```python
#泊松（poisson）分布
#参考：http://www.ruanyifeng.com/blog/2013/01/poisson_distribution.html
rate = 2 #平均发生次数
n = 10
k = np.arange(0, n)
rv = stats.poisson(rate) #泊松分布
print(rv.pmf(0)) #发生0次的概率
print(rv.pmf(k))

plt.plot(k, rv.pmf(k), 'o-')
plt.title('Piosson: rate=%d' % (rate), fontsize=15)
plt.xlabel('Number of occur')
plt.ylabel('Probability of occur', fontsize=15)
plt.show()
```

```python
#超几何分布hypergeom
#举例：500个产品，5个次品，不放回抽20个产品，抽到x的次品的概率
[M, n, N] = [500, 5, 20]
rv = stats.hypergeom(M, n, N)
x = np.arange(0, n+1)
pmf_n = rv.pmf(x)
print(pmf_n)
#如果抽到两个及以上次品被退回，则退回概率
print("被退回概率：%f" % (1-pmf_n[0]-pmf_n[1]))

plt.plot(x, rv.pmf(x), 'o-')
plt.title('Hypergeom: total=%d bad=%d' % (M, n), fontsize=15)
plt.xlabel('Number of get')
plt.ylabel('Probability of get', fontsize=15)
plt.show()
```
